```python
# This code is copied and modified from https://github.com/gariciodaro/adhd-detector

import wget
import pandas as pd
import requests
import csv
import sys
from sys import platform
import os

# seperator in path windows: "\" Linux: "/"
seperator = "/"
# list of disorders to be downloaded (disorder inital are used)
# HBN: healthy
# MDD: Major Depression Disorder
disorders = ['HBN', 'MDD']

# path to id file, where the id will be used to generate the link to download the data
# Note that the code assumes the id file will be the name of the disorder used above + '_ids.csv' (e.g HBN_ids.csv)
# and the file should be one column with the header id or Id
CSV_SUBJECTS_IDS = ''

# amazon AWS link parts (do not change)
URL_AWS          = 'https://fcp-indi.s3.amazonaws.com/data/Projects/HBN/EEG/'
NEXT_URL_AWS     = '/EEG/raw/csv_format/'

# folder where the datya will be stored
PATH_SIGNALS_CSV = CSV_SUBJECTS_IDS+'original_data/'

# type of file to be downloaded (only three types are possible)
FILES_USE        =  ['RestingState_data.csv']#['RestingState_event.csv']#['RestingState_chanlocs.csv']#


# do not change anything below here
def main():
    for disorder in disorders:
        print("Disorder: "+disorder)
        subjects_id=pd.read_csv(CSV_SUBJECTS_IDS+disorder+'_ids.csv')
        try:
            list_ids=subjects_id.id.values.tolist()
        except:
            list_ids=subjects_id.Id.values.tolist()
        for each_subject in list_ids:
                print('Beginning download of subject:',each_subject)
                download_data=True
                #check that the required files exits.
                #if one of them is missing abort download.
                try:
                    #check that the required files exits.
                    #if one of them is missing abort download.
                    for each_file in FILES_USE:
                        url=(URL_AWS+
                            each_subject+
                            NEXT_URL_AWS+
                            each_file)
                        r = requests.head(url).status_code
                        if r!=200:
                            print('URL not found: ', url)
                            download_data=False
                    #download the files
                    if download_data:
                        for each_file in FILES_USE:
                            url = (
                                URL_AWS +
                                each_subject +
                                NEXT_URL_AWS+
                                each_file)
                            if not os.path.isdir(PATH_SIGNALS_CSV+disorder):
                                os.mkdir(PATH_SIGNALS_CSV+disorder)
                            if not os.path.isfile(PATH_SIGNALS_CSV+disorder+seperator + each_subject+"_" + each_file):
                                wget.download(
                                    url,
                                    PATH_SIGNALS_CSV+disorder+seperator + each_subject+"_" + each_file)
                except:
                    print('aborting dowload of subject: ', each_subject)
if __name__ == "__main__":
    main()
```
